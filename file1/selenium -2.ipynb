```python
import selenium
from selenium import webdriver
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By

# calling the automated chrome
driver = webdriver.Chrome()

# opening the naukri.com on automated page
driver.get('https://www.naukri.com/')

# enter designation 
job_designation = driver.find_element(By.CLASS_NAME, "suggestor-input")
job_designation.send_keys('Data Scientist')

# click search button 
search = driver.find_element(By.CLASS_NAME, "qsbSubmit")
search.click()


# filter location
location_filter = driver.find_element(By.CLASS_NAME, "styles_ellipsis__cvWP1 styles_filterLabel__jRP04")
location_filter.click()

# apply button
button = driver.find_element(By.CLASS_NAME, "styles_filter-apply-btn__MDAUd ")
button.click()


# filter salary 
salary_filter = driver.find_element(By.CLASS_NAME, "styles_ellipsis__cvWP1 styles_filterLabel__jRP04")
salary_filter.click()

# apply button
button = driver.find_element(By.CLASS_NAME, "styles_filter-apply-btn__MDAUd ")
button.click()

job_title = []
job_location = []
company_name = []
experience_required = []

# Scraping the required fields

title_tags = driver.find_elements(By.CLASS_NAME, "title ")
for i in title_tags[0:10]:
    title = i.text
    job_title.append(title)
    
location_tags = driver.find_elements(By.CLASS_NAME, "locWdth")
for i in location_tags[0:10]:
    location = i.text
    job_location.append(location if location else "Not available")  # Append "Not available" if location is empty

# Scraping company name, handle if not found
company_tags = driver.find_elements(By.CLASS_NAME, 'comp-name.mw-25')
for i in company_tags[0:10]:
    company = i.text
    company_name.append(company if company else "Not available")  # Append "Not available" if company name is empty

# Scraping experience required, handle if not found
experience_tags = driver.find_elements(By.CLASS_NAME, "expwdth")
for i in experience_tags[0:10]:
    experience = i.text
    experience_required.append(experience if experience else "Not available")  # Append "Not available" if experience is empty

# Ensure all lists have the same length
max_length = max(len(job_title), len(job_location), len(company_name), len(experience_required))
job_title = job_title[:max_length]
job_location = job_location[:max_length]
company_name = company_name[:max_length]
experience_required = experience_required[:max_length]

# Create DataFrame
df = pd.DataFrame({'Job_Title': job_title, 'Location': job_location, 'Company_Name': company_name, 'Experience_R': experience_required})

print(df)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd

# Initialize Chrome WebDriver
driver = webdriver.Chrome()

# Open Flipkart.com
driver.get('https://www.flipkart.com/')

# Enter "sneakers" in the search field
search_input = driver.find_element(By.XPATH, "//input[@title='Search for products, brands and more']")
search_input.send_keys('sneakers')

# Click search button
search_button = driver.find_element(By.XPATH, "//button[@type='submit']")
search_button.click()

# Scraping data for first 100 sneakers
sneakers_data = []

while len(sneakers_data) < 100:
    # Find all sneaker products on the page
    sneakers = driver.find_elements(By.XPATH, "//div[@class='_1AtVbE']")
    
    # Scrape data for each sneaker
    for sneaker in sneakers:
        if len(sneakers_data) >= 100:
            break
        
        brand = sneaker.find_element(By.XPATH, ".//div[@class='_2WkVRV']").text
        description = sneaker.find_element(By.XPATH, ".//a[@class='IRpwTa']").text
        price = sneaker.find_element(By.XPATH, ".//div[@class='_30jeq3']").text
        
        sneakers_data.append({'Brand': brand, 'Product Description': description, 'Price': price})
    
    # Click on the next page button
    try:
        next_button = driver.find_element(By.XPATH, "//a[@class='_1LKTO3'][contains(text(),'Next')]")
        next_button.click()
    except:
        break

# Create DataFrame
df = pd.DataFrame(sneakers_data[:100])  # Trim to first 100 sneakers if more than 100 were scraped

# Print DataFrame
print(df)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd

# Initialize Chrome WebDriver
driver = webdriver.Chrome()

# Open Amazon.in
driver.get('https://www.amazon.in/')

# Enter "Laptop" in the search field and click search icon
search_input = driver.find_element(By.ID, "twotabsearchtextbox")
search_input.send_keys('Laptop')
search_button = driver.find_element(By.ID, "nav-search-submit-button")
search_button.click()

# Set CPU Type filter to "Intel Core i7"
filter_button = driver.find_element(By.XPATH, "//a[@aria-label='Intel Core i7']")
filter_button.click()

# Wait for the filter to apply
time.sleep(2)

# Scrape data for first 10 laptops
laptops_data = []

laptops = driver.find_elements(By.XPATH, "//div[@data-component-type='s-search-result']")[:10]

for laptop in laptops:
    title = laptop.find_element(By.XPATH, ".//span[@class='a-size-medium a-color-base a-text-normal']").text
    try:
        ratings = laptop.find_element(By.XPATH, ".//span[@class='a-icon-alt']").get_attribute('innerHTML')
    except:
        ratings = 'Not available'
    price = laptop.find_element(By.XPATH, ".//span[@class='a-price-whole']").text
    
    laptops_data.append({'Title': title, 'Ratings': ratings, 'Price': price})

# Create DataFrame
df = pd.DataFrame(laptops_data)

# Print DataFrame
print(df)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd

# Initialize Chrome WebDriver
driver = webdriver.Chrome()

# Open azquotes.com
driver.get('https://www.azquotes.com/')

# Click on "Top Quotes"
top_quotes_link = driver.find_element(By.XPATH, "//a[contains(text(),'Top Quotes')]")
top_quotes_link.click()

# Scrape data for Top 1000 quotes
quotes_data = []

# Loop through pages to scrape quotes
while len(quotes_data) < 1000:
    # Find all quote elements on the page
    quote_elements = driver.find_elements(By.XPATH, "//div[@class='wrap-block']")

    # Scrape data for each quote
    for quote_element in quote_elements:
        if len(quotes_data) >= 1000:
            break
        
        # Extract quote, author, and type
        quote_text = quote_element.find_element(By.CLASS_NAME, "title").text
        author_name = quote_element.find_element(By.CLASS_NAME, "author").text
        type_of_quote = quote_element.find_element(By.CLASS_NAME, "kw-box").text
        
        # Append data to quotes_data list
        quotes_data.append({'Quote': quote_text, 'Author': author_name, 'Type Of Quote': type_of_quote})

    # Check if there is a next page and navigate to it
    next_page_link = driver.find_element(By.XPATH, "//a[contains(text(),'Next')]")
    if next_page_link.get_attribute('href'):
        next_page_link.click()
    else:
        break

# Create DataFrame
df = pd.DataFrame(quotes_data)

# Print DataFrame
print(df)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd

# Initialize Chrome WebDriver
driver = webdriver.Chrome()

# Open the webpage
driver.get('https://www.jagranjosh.com/general-knowledge/list-of-all-prime-ministers-of-india-1473165149-1')

# Scraping data for former Prime Ministers
prime_ministers_data = []

# Find all rows containing prime ministers' data
rows = driver.find_elements(By.XPATH, "//table[@class='table4']/tbody/tr")

# Iterate over each row and extract data
for row in rows:
    columns = row.find_elements(By.TAG_NAME, "td")
    
    # Extracting data from columns
    name = columns[0].text
    born_dead = columns[1].text
    term_of_office = columns[2].text
    remarks = columns[3].text
    
    # Append data to prime_ministers_data list
    prime_ministers_data.append({'Name': name, 'Born-Dead': born_dead, 'Term of Office': term_of_office, 'Remarks': remarks})

# Create DataFrame
df = pd.DataFrame(prime_ministers_data)

# Print DataFrame
print(df)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

# Initialize Chrome WebDriver
driver = webdriver.Chrome()

# Open the webpage
driver.get('https://www.motor1.com/')

try:
    # Type in the search bar '50 most expensive cars' and click the search button
    search_bar = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//input[@name='search']")))
    search_bar.send_keys('50 most expensive cars')
    
    search_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "//button[@class='s-form__submit']")))
    search_button.click()
    
    # Click on '50 most expensive cars in the world' link
    expensive_cars_link = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "//a[contains(text(),'50 Most Expensive Cars in the World')]")))
    expensive_cars_link.click()
    
    # Scraping data for the 50 most expensive cars
    expensive_cars_data = []

    # Wait for rows containing car data to be loaded
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//div[@class='s-latest-article']")))
    
    # Find all rows containing car data
    rows = driver.find_elements(By.XPATH, "//div[@class='s-latest-article']")
    
    # Iterate over each row and extract data
    for row in rows:
        car_name = row.find_element(By.CLASS_NAME, "s-latest-article__title").text
        price = row.find_element(By.CLASS_NAME, "s-latest-article__subtitle").text
        
        # Append data to expensive_cars_data list
        expensive_cars_data.append({'Car Name': car_name, 'Price': price})
    
    # Create DataFrame
    df = pd.DataFrame(expensive_cars_data)
    
    # Print DataFrame
    print(df)

except Exception as e:
    print("An error occurred:", e)

finally:
    # Close the browser
    driver.quit()
```
